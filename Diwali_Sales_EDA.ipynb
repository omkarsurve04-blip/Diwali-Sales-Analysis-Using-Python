**Exploratory Data Analysis**

**Gender**

```python
df.columns
```

```python
# plotting a bar chart for Gender and it's count

ax= sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

sales_gen=df.groupby(['Gender'],  as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Gender', y= 'Amount', data= sales_gen)

plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

**Age**

```python
df.columns
```

```python
## plotting a bar chart for Age group with respect to Gender 

ax=sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# Total Amount vs Age Group

sales_age= df.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Age Group', y='Amount', data= sales_age)

plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

**State**

```python
df.columns
```

```python
# total number of orders from top 10 states

orders_state= df.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(7)

sns.set(rc={'figure.figsize':(20,6)})

sns.barplot(x= 'State', y='Orders', data= orders_state)

plt.show()
```

```python
# total amount/sales from top 10 states

sales_states= df.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)

sns.set(rc={'figure.figsize':(20,6)})

sns.barplot(x= 'State', y='Amount', data= sales_states)

plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

**Marital Status**

```python
df.columns
```

```python
ax=sns.countplot(data=df, x='martial_status')

sns.set(rc={'figure.figsize':(7,8)})

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
m_status= df.groupby(['martial_status', 'Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(x= 'martial_status', y='Amount',hue='Gender', data= m_status)

plt.show()
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')

plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

**Product Category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')

plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(7)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')

plt.show()
```

```python
# top 07 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(7).sort_values(ascending=False).plot(kind='bar')

plt.show()
```

**Conclusion**

The Diwali Sales data analysis reveals clear customer purchasing patterns driven by age group, gender, occupation. Married women in the 26–35 age group emerged as the highest contributors to overall sales, while IT, Healthcare, and Aviation professionals showed strong spending behavior. Product categories like Clothing, Food, and Electronics generated the maximum revenue during the festive season.

These insights help businesses optimize targeted marketing, inventory planning, and customer segmentation, enabling more effective sales strategies and improved profitability during peak festive periods.
